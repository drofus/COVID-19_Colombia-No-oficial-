```
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.optimize import minimize
#%matplotlib inline

y = (1 , 3, 9, 16, 24, 45, 54, 65, 93, 108) #número de casos confirmados en Colombia
x = (6, 9, 11, 13, 14, 15, 16, 17, 18, 19)

plt.plot(x,y,'o')
plt.title('Todos los casos de COVID-19 en colombia confirmados')
plt.xlabel('Días del mes de marzo')
plt.ylabel('Casos')
plt.grid()
plt.show()
```

```
y = np.array(y)
x = np.array(x)

log_x = np.log(x)
log_y = np.log(y)

curve_fit = np.polyfit(x, log_y, 1)
print(curve_fit)
```

```
y_ajus = np.exp(-2.08)*np.exp(0.37*x)

plt.plot(x,y,'o', label = 'datos')
plt.plot(x,y_ajus,'r-', label = 'ajuste')
plt.title('Todos los casos de COVID-19 en colombia confirmados')
plt.xlabel('Días del mes de marzo')
plt.ylabel('Casos')
plt.legend()
plt.grid()
plt.show()
```
